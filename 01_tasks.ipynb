# Matplotlib

```python
import matplotlib.pyplot as plt 
```

## Task 1
Using the data frame created above, compute the average salary per department and create a bar chart.
- Rotate x-axis labels.
- Add value labels on top of bars.

```python
# To Do
avg_salary = df.groupby('department')['salary'].mean()

print(avg_salary)
plt.bar(avg_salary.index, avg_salary.values)
plt.xticks(rotation=45)

for i, v in enumerate(avg_salary.values):
    plt.text(i, v, str(int(v)), ha='center', va='bottom')

plt.show()
```

## Task 2 (optional)
Create a scatter plot in which you plot experience on the x-axis and salary on the y-axis.
- Color points by department (different color per department).
- Add an appropriate title, axis labels and a legend.
- Increase the marker size in comparison to the default.

```python
# To Do
departments = df['department'].unique()
colors = ['blue', 'red', 'green']

for dept, color in zip(departments, colors):
    mask = df['department'] == dept
    plt.scatter(df.loc[mask, 'experience'], df.loc[mask, 'salary'], 
                color=color, label=dept, s=100)


plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.show()
```
